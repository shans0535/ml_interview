```python
categorical_columns = ['CpuType_first', 'Connectiontype_first', 'OperatingSystem_first', 'Systemmodel_first']
label_encoders = {col: LabelEncoder() for col in categorical_columns}
for col in categorical_columns:
    final[col + '_encoded'] = label_encoders[col].fit_transform(final[col])
```

```python
categorical_features = final[[col + '_encoded' for col in categorical_columns]].values
combined_features = scipy.sparse.hstack([numerical_features, tfidf_features, categorical_features])
```

```python
combined_features=combined_features.tocsr()
```

```python
# streamlit_app.py
import streamlit as st
import pandas as pd
import numpy as np
import re
import pickle
import scipy
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

# # Load saved models and encoders
# model = pickle.load(open('refresh_model.pkl', 'rb'))
# scaler = pickle.load(open('scaler.pkl', 'rb'))
# tfidf = pickle.load(open('tfidf_vectorizer.pkl', 'rb'))
# label_encoders = pickle.load(open('label_encoders.pkl', 'rb'))  # Dictionary of encoders

# Text extraction helper
def extract_before_after_was(text):
    matches = re.findall(r'([^,]+?)\s+was\s+([^,]+?)(?:,|$)', text)
    before_was = ' '.join([match[0].strip() for match in matches])
    after_was = ' '.join([match[1].strip() for match in matches])
    return before_was, after_was

# Aggregation logic
def aggregate_data(data):
    agg_dict = {
    'RamMB': 'first', 
    'AgentVersion':'first',
    'Manufacturer':'first',
    'DeviceType':'first',
    'CpuType':'first',
    'Connectiontype':'first',
    'OperatingSystem':'first',
    'Systemmodel':'first',
    'MemoryUsagePercent':['mean','max','min'],
    'MemoryUsageMegabytes':['mean','max','min'],
   # 'MemoryPageFileUsagePercent':'mean',
    'Perf_SoftwareHangs': 'sum', 
    'OsCrashes':'sum',
    'SoftwareCrashes':'sum',
    'NetworkUtilizationPercent':'mean',
    'OsServiceFailures':'sum',
    'Description': lambda x: ' | '.join(np.unique(x)),
    'NetworkPacketsReceivedErrors': 'max', 
    'NetworkPacketsOutboundErrors': 'mean',  
    'NetworkPacketRate': 'mean', 
    'DiskSystemFreeSpaceMegabytes':'mean'
    }

    aggregated = data.groupby('SerialNumber').agg(agg_dict).reset_index()
    aggregated.columns = ['_'.join(col) if isinstance(col, tuple) else col for col in aggregated.columns]

    aggregated[['before_was', 'after_was']] = aggregated['description_<lambda>'].apply(
        lambda x: pd.Series(extract_before_after_was(x))
    )
    aggregated['description_combined'] = aggregated['before_was'] + ' ' + aggregated['after_was']

    return aggregated

# Feature engineering
def prepare_features(aggregated):
    # Encode categoricals using saved label encoders
    for col in ['CpuType_first', 'Connectiontype_first', 'OperatingSystem_first', 'Systemmodel_first']:
        le = label_encoders[col]
        aggregated[col + '_encoded'] = le.transform(aggregated[col])

    # TF-IDF from combined text column
    tfidf_features = tfidf.transform(aggregated['description_combined'])

    # Drop non-numeric and text columns to extract numeric ones
    drop_cols = [
        'serial_number',
        'description_<lambda>',
        'before_was',
        'after_was',
        'description_combined',
        'CpuType_first', 'Connectiontype_first', 'OperatingSystem_first', 'Systemmodel_first']
    encoded_cols = [col + '_encoded' for col in label_encoders.keys()]
    
    numerical_features = aggregated.drop(columns=drop_cols + encoded_cols, errors='ignore')
    numerical_scaled = scaler.transform(numerical_features)

    # Combine numeric, tf-idf, and encoded categorical features
    categorical_matrix = scipy.sparse.csr_matrix(aggregated[encoded_cols].values)
    combined = scipy.sparse.hstack([numerical_scaled, tfidf_features, categorical_matrix])

    return combined
# Streamlit UI
st.title("Laptop Refresh Prediction App")

uploaded_file = st.file_uploader("Upload Raw Excel File", type=['xlsx'])

if uploaded_file:
    raw_data = pd.read_excel(uploaded_file)
    st.write("Raw Data Sample", raw_data.head())

    try:
        aggregated = aggregate_data(raw_data)
        st.write("Aggregated Data Sample", aggregated.head())

        features = prepare_features(aggregated)
        predictions = model.predict(features)

        aggregated['Refresh_Prediction'] = predictions
        st.write("Predictions", aggregated[['serial_number', 'Refresh_Prediction']])
        
        st.download_button("Download Predictions as CSV", data=aggregated.to_csv(index=False), file_name='refresh_predictions.csv')
    
    except Exception as e:
        st.error(f"An error occurred: {e}")
```
